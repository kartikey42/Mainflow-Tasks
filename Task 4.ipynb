# Exploratory Data Analysis

```python
import pandas as pd
```

## Load The Data

```python
# Load the Dataset
df = pd.read_csv('USVideos.csv')
```

```python
# Display the First Few Rows of the Data Frame
print(df.head())
```

```python
# Display Basic Information about the Dataset
print(df.info())
```

```python
# Summary Statistics of Numerical Columns
print(df.describe())
```

```python
# Check for Missing Values
print(df.isnull().sum())
```

```python
# Drop Rows with Missing Values (If Any)
df = df.dropna()
```

```python
# Check for Missing Values
print(df.isnull().sum())
```

```python
# Convert 'publish_time' to Datetime
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

## Distribution of Variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the Aesthetic Style of the Plots
sns.set(style='whitegrid')
```

```python
# Distribution of Views
plt.figure(figsize=(10, 6))
sns.histplot(df['views'], bins=50, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Likes
plt.figure(figsize=(10, 6))
sns.histplot(df['likes'], bins=50, kde=True)
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Dislikes
plt.figure(figsize=(10, 6))
sns.histplot(df['dislikes'], bins=50, kde=True)
plt.title('Distribution of Dislikes')
plt.xlabel('Dislikes')
plt.ylabel('Frequency')
plt.show()
```

## Outlier Detection

```python
# Box Plot for Views
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['views'])
plt.title('Box Plot of Views')
plt.xlabel('Views')
plt.show()
```

```python
# Box Plot for Likes
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['likes'])
plt.title('Box Plot of Likes')
plt.xlabel('Likes')
plt.show()
```

```python
# Box Plot for Dislikes
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['dislikes'])
plt.title('Box Plot of Dislikes')
plt.xlabel('Dislikes')
plt.show()
```

## Correlation Analysis

```python
# Compute Correlation Matrix
corr = df[['views', 'likes', 'dislikes', 'comment_count']].corr()
```

```python
# Generate a Heatmap of the Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

## Additional Analysis

```python
# Scatter Plot for Views Vs. Likes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='likes', data=df)
plt.title('Views vs. Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
# Scatter Plot for Views Vs. Dislikes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='dislikes', data=df)
plt.title('Views vs. Dislikes')
plt.xlabel('Views')
plt.ylabel('Dislikes')
plt.show()
```

## Time-Based Analysis

```python
# Extract Year and Month from 'publish_time'
df['year'] = df['publish_time'].dt.year
df['month'] = df['publish_time'].dt.month
```

```python
# Plot the Number of Videos Published Per Year
plt.figure(figsize=(10, 6))
sns.countplot(x='year', data=df)
plt.title('Number of Videos Published per Year')
plt.xlabel('Year')
plt.ylabel('Number of Videos')
plt.show()
```

```python
# Plot the Number of Videos Published Per Month
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=df)
plt.title('Number of Videos Published per Month')
plt.xlabel('Month')
plt.ylabel('Number of Videos')
plt.show()
```
